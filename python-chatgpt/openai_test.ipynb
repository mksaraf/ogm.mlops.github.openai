```python
## Chat Completion
completion = openai.ChatCompletion.create(
  model="gpt-3.5-turbo",
  messages=[
    {"role": "system", "content": "You are a helpful assistant."},
    {"role": "user", "content": "Who is PM of India!"}
  ]
)

print(completion.choices[0].message)
```
